```python
import numpy as np
from sklearn.metrics import f1_score
import pandas as pd
```

```python
def trimf(x, a, b, c):
    #trimf(tinggi=70, a=50, b=65, c=90)
    if x == b:
        return 1.0
    elif x < a or x > c:
        return 0.0
    elif x < b:
        return (x - a) / (b - a) if b != a else 0  #rumus persamaan garis lurus untuk menghitung nilai di antara dua titik dalam garis naik pada fungsi keanggotaan segitiga fuzzy.
    else:
        return (c - x) / (c - b) if c != b else 0

def trapmf(x, a, b, c, d):
    if x < a or x > d:
        return 0.0
    elif b <= x <= c:
        return 1.0
    elif a <= x < b:
        return (x - a) / (b - a) if b != a else 0
    else:  # c < x <= d
        return (d - x) / (d - c) if d != c else 0
```

```python
def fuzzy_mamdani_7(berat, tinggi):
    tinggi_pendek = trimf(tinggi, 50, 50, 82)
    tinggi_sedang = trimf(tinggi, 65, 90, 125)
    tinggi_tinggi = trapmf(tinggi, 110, 125, 200, 200)

    berat_pendek = trimf(berat, 10, 10, 20)
    berat_sedang = trimf(berat, 15, 19, 24)
    berat_tinggi = trapmf(berat, 22, 28, 36, 36)

    rules = [
        (min(berat_pendek, tinggi_pendek), 'kurang'),
        (min(berat_pendek, tinggi_sedang), 'kurang'),
        (min(berat_pendek, tinggi_tinggi), 'kurang'),
        (min(berat_sedang, tinggi_pendek), 'kurang'),
        (min(berat_sedang, tinggi_sedang), 'baik'),
        (min(berat_sedang, tinggi_tinggi), 'baik'),
        (min(berat_tinggi, tinggi_pendek), 'kurang'),
        (min(berat_tinggi, tinggi_sedang), 'baik'),
        (min(berat_tinggi, tinggi_tinggi), 'sangat_baik')
    ]

    universe = np.linspace(0, 3, 100)
    output_kurang = [trimf(u, 0, 0, 1.4) for u in universe]
    output_baik = [trimf(u, 1, 1.8, 2.5) for u in universe]
    output_sangat_baik = [trimf(u, 2, 3, 3) for u in universe]

    aggregated = np.zeros_like(universe)
    for strength, label in rules:
        if strength == 0:
            continue
        if label == 'kurang':
            fuzzy_set = [min(strength, val) for val in output_kurang]
        elif label == 'baik':
            fuzzy_set = [min(strength, val) for val in output_baik]
        elif label == 'sangat_baik':
            fuzzy_set = [min(strength, val) for val in output_sangat_baik]
        aggregated = np.maximum(aggregated, fuzzy_set)

    numerator = np.sum(universe * aggregated)
    denominator = np.sum(aggregated)
    return numerator / denominator if denominator != 0 else 0

```

```python
def to_label_mamdani(val):
    if val < 1.5:
        return 1
    elif val < 2.5:
        return 2
    else:
        return 3

def to_label_sugeno(val):
    if val < 1.5:
        return 1
    elif val < 2.5:
        return 2
    else:
        return 3
```

```python
def main():

    df = pd.read_csv("Child nutrition Data.csv")
    berat = df['berat'].to_numpy()
    tinggi = df['tinggi'].to_numpy()
    umur = df['usia'].to_numpy()
    status = df['status'].to_numpy()
    loop = len(berat)

    mamdani = np.zeros(loop)
    sugeno = np.zeros(loop)
 
    for i in range(loop):
        if umur[i] == 7:
            mamdani_output = fuzzy_mamdani_7(berat[i], tinggi[i])
            sugeno_output = fuzzy_sugeno_7(berat[i], tinggi[i])
            print(f"Mamdani: {mamdani_output:.3f}; Sugeno: {sugeno_output:.3f}; Status: {status[i]}; umur: 7")
            mamdani[i] = mamdani_output
            sugeno[i] = sugeno_output
        elif umur[i] == 8:
            mamdani_output = fuzzy_mamdani_8(berat[i], tinggi[i])
            sugeno_output = fuzzy_sugeno_8(berat[i], tinggi[i])
            print(f"Mamdani: {mamdani_output:.3f}; Sugeno: {sugeno_output:.3f}; Status: {status[i]}; umur: 8")
            mamdani[i] = mamdani_output
            sugeno[i] = sugeno_output
        elif umur[i] == 9:
            mamdani_output = fuzzy_mamdani_9(berat[i], tinggi[i])
            sugeno_output = fuzzy_sugeno_9(berat[i], tinggi[i])
            print(f"Mamdani: {mamdani_output:.3f}; Sugeno: {sugeno_output:.3f}; Status: {status[i]}; umur: 9")
            mamdani[i] = mamdani_output
            sugeno[i] = sugeno_output
        elif umur[i] == 10:
            mamdani_output = fuzzy_mamdani_10(berat[i], tinggi[i])
            sugeno_output = fuzzy_sugeno_10(berat[i], tinggi[i])
            print(f"Mamdani: {mamdani_output:.3f}; Sugeno: {sugeno_output:.3f}; Status: {status[i]}; umur: 10")
            mamdani[i] = mamdani_output
            sugeno[i] = sugeno_output

    mamdani_pred = [to_label_mamdani(x) for x in mamdani]
    sugeno_pred = [to_label_sugeno(x) for x in sugeno]

    f1_mamdani = f1_score(status, mamdani_pred, average='weighted')
    f1_sugeno = f1_score(status, sugeno_pred, average='weighted')

    print(f"F1 Score Mamdani: {f1_mamdani:.3f}")
    print(f"F1 Score Sugeno: {f1_sugeno:.3f}")

if __name__ == "__main__":
    main()
```
